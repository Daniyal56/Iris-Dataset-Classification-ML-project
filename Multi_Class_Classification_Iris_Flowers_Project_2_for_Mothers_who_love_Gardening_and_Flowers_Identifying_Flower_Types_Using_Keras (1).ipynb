```
import numpy 
import pandas 
from keras.models import Sequential 
from keras.layers import Dense 
from keras.utils import to_categorical 
import tensorflow as tf
import keras
```

```
# fix random seed for reproducibility 

seed = 7 
numpy.random.seed(seed) 
```

```
# load dataset 

dataframe = pandas.read_csv("http://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data  ", header=None) 
dataset = dataframe.values 
X = dataset[:,0:4].astype(float) 
Y = dataset[:,4] 
```

```
dataframe[4] = [0 if i=='Iris-setosa' else 1 if i == 'Iris-versicolor' else 2 for i in dataframe[4]]
```

```
y = to_categorical(dataframe[4])
```

```
y
```

```
#creating model

model = Sequential()
model.add(Dense(8, activation='relu', input_shape=(4,)))
model.add(Dense(3, activation='softmax'))
```

```
model.compile(optimizer='Adam', loss='categorical_crossentropy', metrics=['acc'])
```

```
model.fit(X,y,epochs=200)

```

```
#smaller Network

model = Sequential()
model.add(Dense(4, activation='relu', input_shape=(4,)))
model.add(Dense(3, activation='softmax'))

model.compile(optimizer='Adam', loss='categorical_crossentropy', metrics=['acc'])
model.fit(X,y,epochs=200)

```

```
#larger Network

model = Sequential()
model.add(Dense(10, activation='relu', input_shape=(4,)))
model.add(Dense(3, activation='softmax'))

model.compile(optimizer='Adam', loss='categorical_crossentropy', metrics=['acc'])
model.fit(X,y,epochs=200)

```

```
#Overfitting Network

model = Sequential()
model.add(Dense(512, activation='relu', input_shape=(4,)))
model.add(Dense(256, activation='relu'))
model.add(Dense(128, activation='relu'))
model.add(Dense(64, activation='relu'))
model.add(Dense(32, activation='relu'))
model.add(Dense(3, activation='softmax'))

model.compile(optimizer='Adam', loss='categorical_crossentropy', metrics=['acc'])
model.fit(X,y,epochs=1000)

```

```
#Tunning Network

model = Sequential()
model.add(Dense(8, activation='relu', input_shape=(4,)))
model.add(Dense(3, activation='softmax'))

model.compile(optimizer='Adam', loss='categorical_crossentropy', metrics=['acc'])
model.fit(X,y,epochs=250)

```
