```
import tensorflow as tf
import pandas
import keras
from keras.utils import to_categorical
from keras.layers import Dense
```

```
# load dataset 

dataframe = pandas.read_csv("http://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data  ", header=None) 
dataset = dataframe.values 
X = dataset[:,0:4].astype(float) 
Y = dataset[:,4] 
```

```
dataframe[4] = [0 if i == 'Iris-setosa' else 1 if i == 'Iris-versicolor' else 2 for i in dataframe[4]]
```

```
y = to_categorical(dataframe[4])
y
```

```
inputs = keras.Input(shape=(4,))
x = Dense(8, activation='relu')(inputs)
outputs = Dense(3,activation='softmax')(x)
```

```
model = keras.Model(inputs,outputs)
```

```
model.compile(optimizer='Adam', loss='categorical_crossentropy', metrics=['acc'])
```

```
model.fit(X,y, epochs=200)
```

```
#Step 5. Tuning Layers and Number of Neurons in The Model 
#Evaluate a Smaller Network

inputs = keras.Input(shape=(4,))
x = Dense(4, activation='relu')(inputs)
outputs = Dense(3, activation='softmax')(x)

model = keras.Model(inputs,outputs)
model.compile(optimizer='Adam', loss='categorical_crossentropy', metrics=['acc'])
model.fit(X,y,epochs=100)
```

```
#Evaluate a Larger Network

inputs= keras.Input(shape=(4,))
x = Dense(10, activation='relu')(inputs)
outputs = Dense(3,activation='softmax')(x)

#Model

model = keras.Model(inputs,outputs)
model.compile(optimizer='Adam', loss = 'categorical_crossentropy', metrics=['acc'])
model.fit(X,y,epochs=100)
```

```
#Overfitting

inputs = keras.Input(shape=(4,))
x = Dense(512, activation='relu')(inputs)
x = Dense(256, activation='relu')(x)
x = Dense(128, activation='relu')(x)
x = Dense(64, activation='relu')(x)
x = Dense(32, activation='relu')(x)
outputs = Dense(3, activation='softmax')(x)

#model
model = keras.Model(inputs,outputs)
model.compile(optimizer='Adam', loss='categorical_crossentropy', metrics=['acc'])
model.fit(X,y,epochs=1000)
```

```
#Tunning

inputs= keras.Input(shape=(4,))
x = Dense(8, activation='relu')(inputs)
outputs = Dense(3,activation='softmax')(x)

#Model

model = keras.Model(inputs,outputs)
model.compile(optimizer='Adam', loss = 'categorical_crossentropy', metrics=['acc'])
model.fit(X,y,epochs=70)
 
```
